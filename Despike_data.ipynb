```python
from nipype.interfaces import afni as afni


```

```python
for sub_id in range(37):
    
    # Format participant number.
    current_sub = "sub-" + str(sub_id + 1).zfill(2)
    current_num = str(sub_id + 1).zfill(2)

    # Define input and output files.
    current_sub_r1_path = "/Users/la/Documents/Datasets/MovieData_BIDS_preproc/" + current_sub + "/func/" + current_sub + "_task-movie_run-01_space-MNI152NLin2009cAsym_desc-preproc_bold.nii.gz"
    current_sub_r2_path = "/Users/la/Documents/Datasets/MovieData_BIDS_preproc/" + current_sub + "/func/" + current_sub + "_task-movie_run-02_space-MNI152NLin2009cAsym_desc-preproc_bold.nii.gz"

    current_sub_r1_out = "/Users/la/Documents/Datasets/MovieData_BIDS_preproc/" + current_sub + "/func/" + current_sub + "_task-movie_run-01_space-MNI152NLin2009cAsym_desc-preproc_bold_DESPIKE.nii.gz"
    current_sub_r2_out = "/Users/la/Documents/Datasets/MovieData_BIDS_preproc/" + current_sub + "/func/" + current_sub + "_task-movie_run-02_space-MNI152NLin2009cAsym_desc-preproc_bold_DESPIKE.nii.gz"

    # Despike run 1.
    despike_r1 = afni.Despike()
    despike_r1.inputs.in_file = current_sub_r1_path
    despike_r1.inputs.out_file = current_sub_r1_out
    despike_r1.outputtype = 'NIFTI_GZ'
    despike_r1.run()

    # Despike run 2.
    despike_r2 = afni.Despike()
    despike_r2.inputs.in_file = current_sub_r2_path
    despike_r2.inputs.out_file = current_sub_r2_out
    despike_r2.outputtype = 'NIFTI_GZ'
    despike_r2.run()
    
    print("Done processing " + current_sub + ".")

print("All done!")
```
